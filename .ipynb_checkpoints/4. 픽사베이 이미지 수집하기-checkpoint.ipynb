# 1. 이미지 수집하기
* [픽사베이](https://pixabay.com/ko/)

```python
import chromedriver_autoinstaller
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from urllib.request import Request, urlopen
import os
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)

url = 'https://pixabay.com/ko/images/search/고양이/'

driver.get(url)
time.sleep(3)
```

```python
img_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div[4]/div[9]/div/a/img'
img_url = driver.find_element(By.XPATH, img_xpath).get_attribute('src')
print(img_url)
```

```python
img_byte = Request(img_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
f = open('cat.jpg', 'wb')
f.write(urlopen(img_byte).read())
f.close()
```

# 2. 여러개 이미지 수집하기

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)

url = 'https://pixabay.com/ko/images/search/고양이/'

driver.get(url)
time.sleep(3)

img_area_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div'
img_area = driver.find_element(By.XPATH, img_area_xpath)
img_elements = img_area.find_elements(By.TAG_NAME, 'img')

for img in img_elements:
    if img.get_attribute('src') == 'https://pixabay.com/static/img/blank.gif':
        print('blank')
    else:
        img_url = img.get_attribute('src')
        print(img_url)
```

# 3. 함수로 리팩토링
* crawl_image(keyword, pages)

```python
def crawl_image(keyword, pages):
    driver = webdriver.Chrome()
    driver.implicitly_wait(3)
    url = f'https://pixabay.com/ko/images/search/{keyword}/pagi={pages}'
    driver.get(url)
    time.sleep(3)
    folder_name = "keyword"
    os.mkdir(keyword)

    image_urls = []

    image_area_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div'
    image_area = driver.find_element(By.XPATH, image_area_xpath)
    image_elements = image_area.find_elements(By.TAG_NAME, 'img')

    for image_element in image_elements:
        image_url = image_element.get_attribute('data-lazy-src')
        if image_element.get_attribute('data-lazy-src') is None:
            image_url = image_element.get_attribute('src')
        image_urls.append(image_url)

    for i in range(len(image_urls)):
        image_url = image_urls[i]
        image_byte = Request(image_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
                
        f = open(f'sea{i}.jpg', 'wb')
        f.write(urlopen(image_byte).read())
        f.close()
```

```python
def crawl_image(keyword, pages):
    image_urls = []
    for i in range(1, pages+1):
        url = f'https://pixabay.com/ko/images/search/{keyword}/?pagi={i}'
        driver.get(url)
        time.sleep(3)
        image_area_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div'
        image_area = driver.find_element(By.XPATH, image_area_xpath)
        image_elements = image_area.find_elements(By.TAG_NAME, 'img')
        for image_element in image_elements:
            image_url = image_element.get_attribute('data-lazy-src')
            if image_element.get_attribute('data-lazy-src') is None:
                image_url = image_element.get_attribute('src')
            image_urls.append(image_url)
    return image_urls
```

```python
def crawl_and_save_image(keyword, pages):
    image_urls = crawl_image(keyword, pages)
    
    if not os.path.exists(keyword):
        os.mkdir(keyword)
    
    for i in range(len(image_urls)):
        image_url = image_urls[i]
        image_byte = Request(image_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
        filename = image_url.split('/')[-1]
        f = open(f'{keyword}/{filename}', 'wb')
        f.write(urlopen(image_byte).read())
        f.close()
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)

crawl_and_save_image('물고기', 2)
```
